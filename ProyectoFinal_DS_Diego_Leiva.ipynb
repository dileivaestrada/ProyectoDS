### 2.4 쮺칩mo ha evolucionado la poblaci칩n mundial a lo largo del tiempo en cada continente desde 1973 a 2021?

>En el siguiente apartado veremos cu치les son los continentes que han aportado al aumento de la poblaci칩mn mundial y cu치les de estos han experimentado un aumento o disminuci칩n demogr치fica a lo largo del tiempo.

```python
# @title Gr치fico 2: "Poblaci칩n Mundial Total por Continente de 1973 a 2021"

colores_por_region = {
    'Asia': 'green',
    'Americas': 'orange',
    'Africa': 'blue',
    'Europe': 'red',
    'Oceania': 'purple'
}

poblacion_por_continente = df3.groupby(['Year', 'Region'])['PopTotal'].sum().reset_index()
plt.figure(figsize=(12, 8))
for continente in poblacion_por_continente['Region'].unique():
    data_region = poblacion_por_continente[poblacion_por_continente['Region'] == continente]
    plt.plot(data_region['Year'], data_region['PopTotal'], marker='o', linestyle='-', label=continente, color=colores_por_region[continente])
plt.xlabel('A침o')
plt.ylabel('Poblaci칩n Total')
plt.title('Poblaci칩n Mundial Total por Continente de 1973 a 2021')
plt.legend()
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

>**Interpretaci칩n:**

>>En este gr치fico de l칤neas se puede observar la poblaci칩n total por continente, desde 1973 hasta 2021. En este, podemos notar ciertas tendencias y cambios significativos en la distribuci칩n demogr치fica mundial:

### 3.2 Valores perdidos por d칠cada

Analizaremos la cantidad de valores perdidos por d칠cada:

```python
# @title
df3_copia = df3.copy()
df3_copia['Decada'] = (df3_copia['Year'] // 10) * 10

def calcular_estadisticas_nulos_no_nulos(x):
    total_nulos = x.isnull().sum().sum()
    total_no_nulos = x.count().sum()
    total = total_nulos + total_no_nulos
    porcentaje_nulos = (total_nulos / total) * 100 if total > 0 else 0
    return pd.Series({
        'Total Nulos': total_nulos,
        'Total No Nulos': total_no_nulos,
        'Total': total,
        'Porcentaje Nulos': porcentaje_nulos
    })

estadisticas_por_decada = df3_copia.groupby('Decada').apply(calcular_estadisticas_nulos_no_nulos)
estadisticas_por_decada = estadisticas_por_decada.sort_values(by='Porcentaje Nulos', ascending=False)
print("Estad칤sticas de valores nulos y no nulos por d칠cada:")
print(estadisticas_por_decada)
```

>La tabla anterior muestra el porcentaje total de valores nulos por d칠cada. Podemos observar que en la d칠cada los 70's hay un mayor porcentaje de valores perdidos (30% aproximadamente). ## **Secci칩n 7: Optimizaci칩n de hiperpar치metros para Regresi칩n Lineal**

> En esta secci칩n, optimizaremos el hiperpar치metro "fit_intercept" de la Regresi칩n Lineal utilizando GridSearchCV. Calcularemos las m칠tricas de rendimiento con los mejores par치metros encontrados y compararemos estos resultados con los obtenidos previamente para evaluar la efectividad de la optimizaci칩n.

```python
# @title Optimizaci칩n de Hiperpar치metros para la Regresi칩n Lineal usando GridSearchCV

# Definir la cuadr칤cula de hiperpar치metros a explorar
param_grid_lr = {
    'fit_intercept': [True, False],  # Indica si se ajusta o no el t칠rmino independiente
}

# Crear una instancia del modelo de regresi칩n lineal
lr = LinearRegression()

# Crear un objeto GridSearchCV para la regresi칩n lineal
grid_search_lr = GridSearchCV(estimator=lr, param_grid=param_grid_lr, cv=10, scoring='neg_mean_absolute_error')

# Entrenar el modelo con la b칰squeda de cuadr칤cula
grid_search_lr.fit(X_train, y_train)

# Obtener los mejores par치metros encontrados por la b칰squeda de cuadr칤cula
best_params_lr = grid_search_lr.best_params_

# Imprimir los mejores par치metros
print("Mejores par치metros encontrados por GridSearchCV para la regresi칩n lineal:")
print(best_params_lr)

# Crear un nuevo modelo de regresi칩n lineal con los mejores par치metros encontrados
best_lr = LinearRegression(**best_params_lr)

# Entrenar el modelo con los mejores par치metros en el conjunto de entrenamiento
best_lr.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred_best_lr = best_lr.predict(X_test)

# Calcular el MAE y RMSE en el conjunto de prueba
mae_test_best_lr = mean_absolute_error(y_test, y_pred_best_lr)
rmse_test_best_lr = mean_squared_error(y_test, y_pred_best_lr, squared=False)

print(f'MAE en conjunto de prueba (Regresi칩n Lineal con mejores par치metros): {mae_test_best_lr}')
print(f'RMSE en conjunto de prueba (Regresi칩n Lineal con mejores par치metros): {rmse_test_best_lr}')
```

La optimizaci칩n del hiperpar치metro "fit_intercept" no cambi칩 los resultados, **ya** que el mejor par치metro encontrado es el valor predeterminado (True), el cual ya fue utiliado cuando se evalu칩 el modelo a trav칠s de Regresi칩n Lineal.

## **Secci칩n 8: Aplicaci칩n de PCA para predecir de la esperanza de vida al nacer (LifeExpBirth)**

> En esta secci칩n, aplicamos el An치lisis de Componentes Principales (PCA) para reducir la dimensionalidad del conjunto de datos. Seleccionamos el n칰mero 칩ptimo de componentes principales y luego entrenamos y evaluamos modelos de Regresi칩n Lineal (LR) y Support Vector Regressor (SVR) utilizando estos componentes reducidos. Finalmente, calculamos y comparamos las m칠tricas de rendimiento de ambos modelos con y sin PCA y sacamos una conclusi칩n general sobre su efectividad.

Primero, veremos nuevamente el Dataframe Imputado:

```python
# @title
df3_nuevo_imputado.info()
```

Podemos observar que existen 5 variables no num칠ricas. Para trabajar con PCA, consideraremos eliminar las columnas categ칩ricas. Llamaremos "df3_pca" al nuevo dataframe sin variables categ칩ricas y que usaremos para aplicar PCA:
:

```python
# @title
# Lista de nombres de columnas categ칩ricas a eliminar
columnas_categoricas = ['Year', 'Country', 'Region', 'SubRegion']

# Crear un nuevo DataFrame sin las columnas categ칩ricas
df3_pca = df3_nuevo_imputado.drop(columns=columnas_categoricas)

# Verificar el nuevo DataFrame
df3_pca.info()
```

Como podemos ver, el dataframe ha quedado con 25 columnas num칠ricas. El paso siguiente ser치 la estandarizaci칩n de las variables.

```python
# @title
# Inicializar el objeto StandardScaler
sc = StandardScaler()

# Estandarizar las variables en df3_pca
df3_pca_sc = sc.fit_transform(df3_pca)

# Crear un nuevo DataFrame con las variables estandarizadas
df3_pca_sc = pd.DataFrame(df3_pca_sc, columns=df3_pca.columns)

# Verificar el nuevo DataFrame estandarizado
df3_pca_sc.head()
```

Ahora, aplicaremos PCA a todo el dataframe estandarizado:

```python
# @title
# Inicializar PCA con el n칰mero deseado de componentes
pca = PCA()

# Aplicar PCA al dataframe estandarizado
pca.fit(df3_pca_sc)

# Obtener la varianza explicada por cada componente principal
explained_variance_ratio = pca.explained_variance_ratio_

# Imprimir la varianza explicada por cada componente principal
print("Varianza explicada por cada componente principal:")
for i, variance_ratio in enumerate(explained_variance_ratio):
    print(f"Componente {i + 1}: {variance_ratio:.4f}")
```

Para poder ver la varianza acumulada, observaremos su comportamiento mediante un gr치fico de l칤neas:

```python
# @title
# Obtener la varianza explicada acumulada
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Graficar la varianza explicada acumulada
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='-')
plt.xlabel('N칰mero de componentes principales')
plt.ylabel('Varianza explicada acumulada')
plt.title('Varianza explicada acumulada por n칰mero de componentes principales')
plt.grid(True)
```

Como se puede observar en el gr치fico, se observa una varianza superior a 0.9 desde las 14 componentes principales hacia arriba. Por lo tanto, se decide utilizar 14 Componentes principales:

```python
# @title
# Inicializar PCA con el n칰mero deseado de componentes
pca = PCA(n_components=14)

# Aplicar PCA al dataframe estandarizado
X_pca = pca.fit_transform(df3_pca_sc)
y = df3_pca_sc['LifeExpBirth']
```

Dividimos en training y test al 30%

```python
# @title
# Dividimos los datos en conjuntos de entrenamiento y prueba
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y, test_size=0.3, random_state=42)
```

Aplicaremos regresi칩n lineal y c치lculo de m칠tricas

```python
# @title
# Regresi칩n Lineal
lr_pca = LinearRegression()
lr_pca.fit(X_train_pca, y_train_pca)
y_pred_lr_pca = lr_pca.predict(X_test_pca)

# Calcular MAE y RMSE
mae_test_lr_pca = mean_absolute_error(y_test_pca, y_pred_lr_pca)
rmse_test_lr_pca = mean_squared_error(y_test_pca, y_pred_lr_pca, squared=False)

# Validaci칩n cruzada con k=10
cv_scores_mae_lr_pca = cross_val_score(lr_pca, X_pca, y, cv=10, scoring='neg_mean_absolute_error')
cv_scores_rmse_lr_pca = cross_val_score(lr_pca, X_pca, y, cv=10, scoring='neg_root_mean_squared_error')

# Calcular el MAE y RMSE promedio
mae_avg_lr_pca = -cv_scores_mae_lr_pca.mean()
rmse_avg_lr_pca = -cv_scores_rmse_lr_pca.mean()

print(f'MAE en conjunto de prueba (Regresi칩n Lineal con PCA): {mae_test_lr_pca:.2f}')
print(f'RMSE en conjunto de prueba (Regresi칩n Lineal con PCA): {rmse_test_lr_pca:.2f}')
print(f'MAE promedio usando 10-fold cross-validation (Regresi칩n Lineal con PCA): {mae_avg_lr_pca:.2f}')
print(f'RMSE promedio usando 10-fold cross-validation (Regresi칩n Lineal con PCA): {rmse_avg_lr_pca:.2f}')
```

Para comparar el desempe침o de el dataframe con PCA versus el desempe침o del dataframe sin PCA, utilizaremos el coeficiente de determinaci칩n R2:

```python
# @title
# SVR
svr_pca = SVR()
svr_pca.fit(X_train_pca, y_train_pca)
y_pred_svr_pca = svr_pca.predict(X_test_pca)

# Calcular MAE y RMSE en el conjunto de prueba
mae_test_svr_pca = mean_absolute_error(y_test_pca, y_pred_svr_pca)
rmse_test_svr_pca = mean_squared_error(y_test_pca, y_pred_svr_pca, squared=False)

# Validaci칩n cruzada con k=10
cv_scores_mae_svr_pca = cross_val_score(svr_pca, X_pca, y, cv=10, scoring='neg_mean_absolute_error')
cv_scores_rmse_svr_pca = cross_val_score(svr_pca, X_pca, y, cv=10, scoring='neg_root_mean_squared_error')

# Calcular el MAE y RMSE promedio
mae_avg_svr_pca = -cv_scores_mae_svr_pca.mean()
rmse_avg_svr_pca = -cv_scores_rmse_svr_pca.mean()

print(f'MAE en conjunto de prueba (SVR con PCA): {mae_test_svr_pca:.2f}')
print(f'RMSE en conjunto de prueba (SVR con PCA): {rmse_test_svr_pca:.2f}')
print(f'MAE promedio usando 10-fold cross-validation (SVR con PCA): {mae_avg_svr_pca:.2f}')
print(f'RMSE promedio usando 10-fold cross-validation (SVR con PCA): {rmse_avg_svr_pca:.2f}')
```

Conclusi칩n:

El R2 entregado por el modelo sin PCA fue de de 0.88, mientras que, al aplicar PCA y estandarizaci칩n se obtuvo un R2 de 0.91. Esto sugiere que el modelo con PCA tiene un rendimiento ligeramente mejor. Con esto, podemos decir que al momento de reducir la dimensionalidad a 14, el modelo adquiere un mejor rendimiento.
