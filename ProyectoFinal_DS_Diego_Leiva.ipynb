### 2.3 Primer Acercamiento a la Evolución Demográfica Global: Análisis de la Población Mundial de 1973 a 2021

>Para conocer el contexto del DataFrame, es importante comenzar por la evolución demográfica que ha tenido el planeta. Se comenzará mostrando la población total del mundo desde 1973 a 2021.

```python
# @title Gráfico 1: "Población Total de Mundo desde 1973 a 2021"
poblacion_mundial = df3.groupby('Year')['PopTotal'].sum().reset_index()
plt.figure(figsize=(12, 8))
plt.plot(poblacion_mundial['Year'], poblacion_mundial['PopTotal'], marker='o', linestyle='-', color='black')
plt.xlabel('Año')
plt.ylabel('Población Mundial Total')
plt.title('Población Mundial Total por Año')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
# @title

poblacion_1973 = df3[df3['Year'] == 1973]['PopTotal'].sum()
print('La población total en 1973 es:', poblacion_1973)
poblacion_2021 = df3[df3['Year'] == 2021]['PopTotal'].sum()
print('La población total en 2021 es:', poblacion_2021)
```



> **Interpretación:**

>>En el presente gráfico de líneas, se puede observar que la población total presenta un aumento sostenido en el transcurso de los años.

>>En el primer año de estudio (1973) podemos observar que había una población total mundial de aproximadamente 3,8 billones de habitantes, mientras que en el último año de estudio (2021) se observa una población total de casi 7,8 billones, lo que indica que, en casi cuatro décadas, se duplicó la cantidad de habitantes en el mundo.

>>Una posible explicación a este fenómeno podría deberse a factores tales como el aumento de la tasa de natalidad a lo largo del tiempo o las  mejoras significativas en la esperanza de vida en diversas regiones del planeta. Asimismo, este crecimiento demográfico podría ser atribuído a factores socioeconómicos, avances tecnológicos y a un acceso más amplio a los servicios de salud.

### 2.4 ¿Cómo ha evolucionado la población mundial a lo largo del tiempo en cada continente desde 1973 a 2021?

>En el siguiente apartado veremos cuáles son los continentes que han aportado al aumento de la poblaciómn mundial y cuáles de estos han experimentado un aumento o disminución demográfica a lo largo del tiempo.

```python
# @title Gráfico 2: "Población Mundial Total por Continente de 1973 a 2021"

colores_por_region = {
    'Asia': 'green',
    'Americas': 'orange',
    'Africa': 'blue',
    'Europe': 'red',
    'Oceania': 'purple'
}

poblacion_por_continente = df3.groupby(['Year', 'Region'])['PopTotal'].sum().reset_index()
plt.figure(figsize=(12, 8))
for continente in poblacion_por_continente['Region'].unique():
    data_region = poblacion_por_continente[poblacion_por_continente['Region'] == continente]
    plt.plot(data_region['Year'], data_region['PopTotal'], marker='o', linestyle='-', label=continente, color=colores_por_region[continente])
plt.xlabel('Año')
plt.ylabel('Población Total')
plt.title('Población Mundial Total por Continente de 1973 a 2021')
plt.legend()
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

>**Interpretación:**

>>En este gráfico de líneas se puede observar la población total por continente, desde 1973 hasta 2021. En este, podemos notar ciertas tendencias y cambios significativos en la distribución demográfica mundial:

>>En primer lugar, Asia se muestra como el continente con mayor cantidad de habitantes durante todo el período de estudio, separándose de forma considerable de las demás regiones y mostrando una clara tendencia al aumento a lo largo de los años. Luego, se puede apreciar que Europa en el año 1973 era el segundo continente con mayor población, seguido de América, África y, finalmente, Oceanía.

>>Para el año 2021, el panorama ha cambiado; En la actuliadad, África es el segundo continente con mayor población, con una tendecia ascendente sostenida a través del tiempo, seguido de América, que se posciona tercero para el 2021, región también muestra un aumento durante estos años de estudio, mientras que Europa se posiciona como cuarto y penúltimo continente en la actualidad a nivel demográfico, manteniendo prácticamente su población en estos ultimos años.

>>En contraste a todo lo anterior, observamos que Oceanía ha sido el continente con menor población durante todos estos años, la cual no ha variado de forma significativa.

### 2.8 Explorando el comportamiento económico: Análisis de la variabilidad del producto interno bruto *(GDP)* por continente en el mundo

>El Producto Interno Bruto (GDP, por sus siglas en inglés) es una medida que cuantifica el valor total de bienes y servicios producidos en un país en un determinado periodo. Es un indicador clave que permite ver el desempeño económico de una nación, pues abarca todos los sectores desde la agricultura, la industria, hasta los servicios básicos. Analizar el PIB y su comportamiento en cada región o país, permitirá obtener una visión de la economía global y su evolución. Se espera que el PIB exhiba las diferencias económicas significativas entre naciones y podría ayudar a conocer aquellos factores que impulsan o limitan su desarrollo según el contexto de cada continente o país.

#### 3.4.1 Tratamiento de outliers e imputación de Datos de columna "FertRate":

```python
# @title Gráfico 11: Boxplot de Tasa de Fertidilidad (FertRate)
sns.boxplot(x=df3_nuevo['FertRate'])
plt.title('Boxplot de Tasa de Fertilidad')
plt.xlabel('FertRate')
plt.show()
```

Como se observa, la columna FertRate tiene puntos atípicos por encima del cuartil 3. Esto quiere decir que existen cifras significativamente mayores. Eso de puede deber a que efectivamente ocurrieron sucesos en algún país/año en donde este número aumentó de forma considerable. Para estos puntos se realizará un tratamiento de outliers.

```python
# @title Imputación de outliers de FertRate por la mediana
q1 = df3_nuevo['FertRate'].quantile(0.25)
q3 = df3_nuevo['FertRate'].quantile(0.75)
iqr = q3 - q1

limite_inferior = q1 - 1.5 * iqr
limite_superior= q3 + 1.5 * iqr

outliers = df3_nuevo[(df3_nuevo['FertRate'] < limite_inferior) | (df3_nuevo['FertRate'] > limite_superior)]
df3_nuevo = df3_nuevo[(df3_nuevo['FertRate'] >= limite_inferior) & (df3_nuevo['FertRate'] <= limite_superior)]

imputador = SimpleImputer(strategy='median')

df3_nuevo['FertRate'] = imputador.fit_transform(df3_nuevo[['FertRate']])
```

Para este caso, se decidió imputar por la mediana y eliminar los outliers de la columna FertRate.

## **Sección 5: Evaluación del modelo para predecir la esperanza de vida al nacer (LifeExpBirth) mediante Regresión Lineal (LR)**

> En esta sección aplicaremos método de regresión lineal al DataFrame, calcularemos sus respectivas métricas, las cuales serán comparadas para su interpretación y conclusión final.

Para la predicción de la Esperanza de vida al nacer (LifeExpBirth) utilizando un modelo de regresión, se adaptarán las variables no numéricas del modelo:

### 5.1 Separación del DataFrame y training test split

```python
# @title Preparación de los datos: Apartamos el Label (LifeExpBirth) y variables no numéricas del DataFrame

df_x = df3_nuevo_imputado.copy()
df_x = df_x.drop(columns= ['LifeExpBirth', 'Year', 'Country', 'Region', 'SubRegion'])
```

```python
# @title Declaración LifeExpBirth como variable independiente

df_y = df3_nuevo_imputado[['LifeExpBirth']]
```

Para comenzar, realizamos la división de conjuntos de entrenamiento y prueba en 70% y 30%, respectivamente

```python
# @title Divisón en conjutos de training y test (70/30)%

X_train,X_test,y_train,y_test = train_test_split(df_x,df_y,test_size=0.3,random_state=42)
```

```python
# @title  Verifica las dimensiones de los conjuntos

print("Dimensiones de X_train:", X_train.shape)
print("Dimensiones de X_test:", X_test.shape)
print("Dimensiones de y_train:", y_train.shape)
print("Dimensiones de y_test:", y_test.shape)
```

## **Sección 8: Aplicación de PCA para predecir de la esperanza de vida al nacer (LifeExpBirth)**

> En esta sección, aplicamos el Análisis de Componentes Principales (PCA) para reducir la dimensionalidad del conjunto de datos. Seleccionamos el número óptimo de componentes principales y luego entrenamos y evaluamos modelos de Regresión Lineal (LR) y Support Vector Regressor (SVR) utilizando estos componentes reducidos. Finalmente, calculamos y comparamos las métricas de rendimiento de ambos modelos con y sin PCA y sacamos una conclusión general sobre su efectividad.

Primero, veremos nuevamente el Dataframe Imputado:

```python
# @title
df3_nuevo_imputado.info()
```

Podemos observar que existen 5 variables no numéricas. Para trabajar con PCA, consideraremos eliminar las columnas categóricas. Llamaremos "df3_pca" al nuevo dataframe sin variables categóricas y que usaremos para aplicar PCA:
:

```python
# @title
# Lista de nombres de columnas categóricas a eliminar
columnas_categoricas = ['Year', 'Country', 'Region', 'SubRegion']

# Crear un nuevo DataFrame sin las columnas categóricas
df3_pca = df3_nuevo_imputado.drop(columns=columnas_categoricas)

# Verificar el nuevo DataFrame
df3_pca.info()
```

Como podemos ver, el dataframe ha quedado con 25 columnas numéricas. El paso siguiente será la estandarización de las variables.

```python
# @title
# Inicializar el objeto StandardScaler
sc = StandardScaler()

# Estandarizar las variables en df3_pca
df3_pca_sc = sc.fit_transform(df3_pca)

# Crear un nuevo DataFrame con las variables estandarizadas
df3_pca_sc = pd.DataFrame(df3_pca_sc, columns=df3_pca.columns)

# Verificar el nuevo DataFrame estandarizado
df3_pca_sc.head()
```

Ahora, aplicaremos PCA a todo el dataframe estandarizado:

```python
# @title
# Inicializar PCA con el número deseado de componentes
pca = PCA()

# Aplicar PCA al dataframe estandarizado
pca.fit(df3_pca_sc)

# Obtener la varianza explicada por cada componente principal
explained_variance_ratio = pca.explained_variance_ratio_

# Imprimir la varianza explicada por cada componente principal
print("Varianza explicada por cada componente principal:")
for i, variance_ratio in enumerate(explained_variance_ratio):
    print(f"Componente {i + 1}: {variance_ratio:.4f}")
```

Para poder ver la varianza acumulada, observaremos su comportamiento mediante un gráfico de líneas:

```python
# @title
# Obtener la varianza explicada acumulada
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Graficar la varianza explicada acumulada
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='-')
plt.xlabel('Número de componentes principales')
plt.ylabel('Varianza explicada acumulada')
plt.title('Varianza explicada acumulada por número de componentes principales')
plt.grid(True)
```

Como se puede observar en el gráfico, se observa una varianza superior a 0.9 desde las 14 componentes principales hacia arriba. Por lo tanto, se decide utilizar 14 Componentes principales:

```python
# @title
# Inicializar PCA con el número deseado de componentes
pca = PCA(n_components=14)

# Aplicar PCA al dataframe estandarizado
X_pca = pca.fit_transform(df3_pca_sc)
y = df3_pca_sc['LifeExpBirth']
```

Dividimos en training y test al 30%

```python
# @title
# Dividimos los datos en conjuntos de entrenamiento y prueba
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y, test_size=0.3, random_state=42)
```

Aplicaremos regresión lineal y cálculo de métricas

```python
# @title
# Regresión Lineal
lr_pca = LinearRegression()
lr_pca.fit(X_train_pca, y_train_pca)
y_pred_lr_pca = lr_pca.predict(X_test_pca)

# Calcular MAE y RMSE
mae_test_lr_pca = mean_absolute_error(y_test_pca, y_pred_lr_pca)
rmse_test_lr_pca = mean_squared_error(y_test_pca, y_pred_lr_pca, squared=False)

# Validación cruzada con k=10
cv_scores_mae_lr_pca = cross_val_score(lr_pca, X_pca, y, cv=10, scoring='neg_mean_absolute_error')
cv_scores_rmse_lr_pca = cross_val_score(lr_pca, X_pca, y, cv=10, scoring='neg_root_mean_squared_error')

# Calcular el MAE y RMSE promedio
mae_avg_lr_pca = -cv_scores_mae_lr_pca.mean()
rmse_avg_lr_pca = -cv_scores_rmse_lr_pca.mean()

print(f'MAE en conjunto de prueba (Regresión Lineal con PCA): {mae_test_lr_pca:.2f}')
print(f'RMSE en conjunto de prueba (Regresión Lineal con PCA): {rmse_test_lr_pca:.2f}')
print(f'MAE promedio usando 10-fold cross-validation (Regresión Lineal con PCA): {mae_avg_lr_pca:.2f}')
print(f'RMSE promedio usando 10-fold cross-validation (Regresión Lineal con PCA): {rmse_avg_lr_pca:.2f}')
```

Para comparar el desempeño de el dataframe con PCA versus el desempeño del dataframe sin PCA, utilizaremos el coeficiente de determinación R2:

```python
# @title
# SVR
svr_pca = SVR()
svr_pca.fit(X_train_pca, y_train_pca)
y_pred_svr_pca = svr_pca.predict(X_test_pca)

# Calcular MAE y RMSE en el conjunto de prueba
mae_test_svr_pca = mean_absolute_error(y_test_pca, y_pred_svr_pca)
rmse_test_svr_pca = mean_squared_error(y_test_pca, y_pred_svr_pca, squared=False)

# Validación cruzada con k=10
cv_scores_mae_svr_pca = cross_val_score(svr_pca, X_pca, y, cv=10, scoring='neg_mean_absolute_error')
cv_scores_rmse_svr_pca = cross_val_score(svr_pca, X_pca, y, cv=10, scoring='neg_root_mean_squared_error')

# Calcular el MAE y RMSE promedio
mae_avg_svr_pca = -cv_scores_mae_svr_pca.mean()
rmse_avg_svr_pca = -cv_scores_rmse_svr_pca.mean()

print(f'MAE en conjunto de prueba (SVR con PCA): {mae_test_svr_pca:.2f}')
print(f'RMSE en conjunto de prueba (SVR con PCA): {rmse_test_svr_pca:.2f}')
print(f'MAE promedio usando 10-fold cross-validation (SVR con PCA): {mae_avg_svr_pca:.2f}')
print(f'RMSE promedio usando 10-fold cross-validation (SVR con PCA): {rmse_avg_svr_pca:.2f}')
```

Conclusión:

El R2 entregado por el modelo sin PCA fue de de 0.88, mientras que, al aplicar PCA y estandarización se obtuvo un R2 de 0.91. Esto sugiere que el modelo con PCA tiene un rendimiento ligeramente mejor. Con esto, podemos decir que al momento de reducir la dimensionalidad a 14, el modelo adquiere un mejor rendimiento.
